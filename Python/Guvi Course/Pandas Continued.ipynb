# Merge 
## Merge will work like JOINS in SQL, be aware

```python
d17 ={
   'ID': ['1','3','6','7','8'],
   'Feature1': ['AB','BC','CD','DE','EF'],
   'Feature2': ['KL','LM','M-N',"N-O",'O-P']
}


d18 = {
   'ID': ['1','2','3','4','5'],
   'Feature1': ['A','B','C','D','E'],
   'Feature2': ['L','M','N',"O",'P']
}
```

```python
df17 = pd.DataFrame(d17)
df18 = pd.DataFrame(d18)
```

### Inner Join

```python
df_merge_o = pd.merge(df17,df18,on="ID",how="inner")
df_merge_o
```

## Right Join

```python
df_merge_r = pd.merge(df17,df18,on="ID",how="right")
df_merge_r
```

## Left Join

```python
df_merge_l = pd.merge(df17,df18,on='ID', how="left")
df_merge_l
```

```python
df_merge_l = pd.merge(df17,df18,on="ID")
df_merge_l
```

## concat
#### can have common or differnt
## join
#### one differnt col atleast
## Merge
#### atleast one common column

## Based on the combinations we get the rows count

```python
df_merge_l = pd.merge(df17,df18, how="cross")
df_merge_l
```

```python
df_merge_l = pd.merge(df17,df18, how="cross")
df_merge_l
```
