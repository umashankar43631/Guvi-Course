# Series  vs List
#### Series will show indexes starting with 0 while list won't

```python
import numpy as np
```

```python
data = np.array(['A','B','C','D'])
s3 = pd.Series(data)
print(s3)
```

```python
data1 = np.array(['a','b','c','d'])
s4 = pd.Series(data, index=[i for i in range(100,104)])
print(s4)
```

### Retrieving more than 1 value from series

```python
s6 = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(s6[['a','b','c']])
```

```python
s6 = pd.Series(['a','b','c','d','e'], index=[1,2,3,4,5])
print(s6[[2,3,4]])
```
