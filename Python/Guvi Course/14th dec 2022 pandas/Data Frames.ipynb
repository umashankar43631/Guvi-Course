### since it can't convert strings to float it will convert who and all can be converted to float to float

```python
data2 = [['Guvi1',10],['Guvi2',12],['Guvi3',13]]
df3 = pd.DataFrame(data2, columns=['Name','Age'], dtype='float64')
print(df3)
```

```python
data2 = [['Guvi1',10],['Guvi2',12],['Guvi3',13]]
df3 = pd.DataFrame(data2, dtype=float,columns=['Name','Age'])
print(df3)
```
