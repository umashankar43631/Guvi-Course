## Pivot Table
### Pivot table is used for aggregated values or min, max, count
### 

## By default it will apply mean

```python
pd.pivot_table(df4, index=['Pclass'])
```

###  In the sex pivot table survived rate is like 74% of females out of 100 and 18% of males out of 100

#### average age of females is 27 and 30 for men

#### Average Fare of females for ticket is 44$ and for men is 25.5$



```python
pd.pivot_table(df4, index=['Sex'])
```

## The Below pivot table sorts the results by 2 indexes
### females with pclass 1 

```python
pd.pivot_table(df4, index=['Sex','Pclass'], aggfunc=['count'])
```

```python
pd.pivot_table(df4, index=['Sex', 'Pclass'], values=['Survived', 'Age'])
```
